#####  Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
 def standard_dev(data): #define standard_dev function that takes data as argument
        variance = [] #empty list thats going to contain variance formula
        sample_mean = np.mean(data) #calculate sample mu 
        variance_1 = [(i - sample_mean)**2 for i in data] # for each i in data, subtract sample mean and square it
        variance_2 = np.sum(variance_1) # and then add all the values together
        variance_3 = variance_2/len(data) #devide sum values by total sample number to get variance
        return variance_3**0.5 # square root variance to get standard deviation
    
```

```python
df_columns = df_new[['ACT Participation %', 'ACT English', 'ACT Math', 'ACT Reading', 'ACT Science', 'ACT Composite', 
                    'SAT Participation %', 'SAT Evidence-Based Reading and Writing', 'SAT Math', 'SAT Total']]
df_columns #renaming columns by dropping state since its not a numerical value
```

```python
sd = []#append location
def std_append(df): #define std_append function that takes df as argument     
    for column in df: #iterating dataframe to get each column
        sd.append(standard_dev(df[column])) #by calling standard_dev functiong, append each calculated std values to list sd
    return sd #append each calculated std values to list sd and return the output of the function
    
std_append(df_columns)
```

## Step 3: Descriptive and Inferential Statistics

##### Summarize each distribution. As data scientists, be sure to back up these summaries with statistics. (Hint: What are the three things we care about when describing distributions?)

1.) Center of distribution (mean,median,mode)

2.) Characteristic of spread from middle point or average/mean (standard deviation, variance, range)

3.) Shape of distribution (Skewness of graph)

If distribution is symmetric, mean == median == mode

If distribution is negatively skewed, mean < median < mode

If distribution is positibely skewed, mode < median < mean

##### Summarize each relationship. Be sure to back up these summaries with statistics.

```python
df_new.set_index('State',inplace=True) #shiftinf state column to index label because state column contains strings
```

```python
def calculate_everything(df): #define calculate_everything function that takes df as argument
    for col in df: #iterate each column in dataframe,
        print("For column", col) #print the column name 
        mode = stats.mode(df[col]).mode #calculate mode using stats library
        mean = np.mean(df[col]) #caculate mean using numpy
        median = np.median(df[col]) #calculate median using numpy
        stand_dev = np.std(df[col]) #calculate using standard deviation using numpy
        print('Mode', mode) #print calculated mode
        print('Median', median) #print calculated median
        print('Mean', mean) #print calculated mean
        print('Standard Deviation', stand_dev) #print calculated standard deviation
        if mean > median: #if mean is greater than median
            print('The data is skewed right/positive') #print this
        elif mean < median: #else if mean is less than median
            print('The data is skewed left/negative') #print this
        else:#everything else,
            print('The data is symmetrically distributed')#print this
        print('\n')#print new line to make it look nice

calculate_everything(df_new) #calling the function using df_new dataframe            
```

##### Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

$H_0: \mu SAT Participation = \mu ACT Participation$

$H_A: \mu SAT Participation \neq \mu ACT Participation$

```python
act_col = df_new['ACT Participation %'] #pulling data from column     
sat_col = df_new['SAT Participation %'] #ACT and SAT participation
                                        #as variable,
stats.ttest_ind(act_col, sat_col)#perform two-sample t-test
```

P_value is 0.0002 which is less than alpha $\alpha = 0.05$. Based on this, I can conclude that it is safe to reject the null hypothesis.

##### Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
stats.t.interval(0.95, 
                 len(df_new['SAT Participation %']) -1, 
                 loc = np.mean(df_new['SAT Participation %']),
                 scale = np.std(df_new['SAT Participation %'], ddof = 1)) #calculatew two t-test
```

Based on the calculation above, I am 95% confident that the true mean of number of SAT participation rate is between -31.051 and 110.659

```python
stats.t.interval(0.95, 
                 len(df_new['ACT Participation %']) -1, 
                 loc = np.mean(df_new['ACT Participation %']),
                 scale = np.std(df_new['ACT Participation %'], ddof = 1))
```

Based on the calculation above, I am 95% confident that the true mean of number of ACT participation rate is between 0.698 and 129.812
